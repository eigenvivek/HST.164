```julia
using Plots; pyplot()
using FFTW
using DSP
```

```julia
# Model parameters
C0 = 440
T2 = 1/8

# Number of points 
N = 2^14 - 1 
Ts = 1 / (1.1 * N) 
t0 = 0 
tmax = t0 + N * Ts
t = t0:Ts:tmax;
```

```julia
# Signal 1
signal = C0 * exp.(-t / T2);
F = fft(signal) |> fftshift
freqs = fftfreq(length(t), 1.0/Ts) |> fftshift

time_domain = plot(t, signal, title = "Signal")
freq_domain = plot(freqs, abs.(F), title = "Spectrum", xlim=(-500, +500)) 
plot(time_domain, freq_domain, layout = 2)
```

```julia
# Signal 1
sigma = 1

t = -tmax:Ts / 2:tmax
signal = C0 * exp.(- 2 .^ t / (4 * sigma^2));
F = fft(signal) |> fftshift
freqs = fftfreq(length(t), 1.0/Ts) |> fftshift

time_domain = plot(t, signal, title = "Signal")
freq_domain = plot(freqs, abs.(F), title = "Spectrum", xlim=(-50, +50)) 
plot(time_domain, freq_domain, layout = 2)
```
